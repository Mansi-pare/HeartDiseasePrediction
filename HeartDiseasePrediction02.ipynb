```python
# Data Preprocessing
X = df.drop('HeartDiseaseorAttack', axis=1)
y = df['HeartDiseaseorAttack']
```

```python
# Handle Class Imbalance
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

```

```python
print(df['HeartDiseaseorAttack'].value_counts())
```

```python
# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)
```

```python
# Feature Scaling
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Initialize Models
models = {
    'Logistic Regression': LogisticRegression(),
    'Random Forest': RandomForestClassifier(n_estimators=100, random_state=42),
    'K-Nearest Neighbors': KNeighborsClassifier(n_neighbors=5)
}
```

```python
# Initialize Models
models = {
    'Logistic Regression': LogisticRegression(),
    'Random Forest': RandomForestClassifier(n_estimators=100, random_state=42),
    'K-Nearest Neighbors': KNeighborsClassifier(n_neighbors=5)
}

# Check if models are initialized
print(models)
```

```python
# Model Training and Evaluation
accuracy_scores = {}
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    accuracy_scores[name] = acc
    print(f'\n{name} Accuracy: {acc:.4f}')
    print(classification_report(y_test, y_pred))
    sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d')
    plt.title(f'{name} Confusion Matrix')
    plt.show()
```

```python
# Compare Model Accuracies
import matplotlib.pyplot as plt
import seaborn as sns

# Sort accuracy scores in decreasing order
sorted_accuracy = dict(sorted(accuracy_scores.items(), key=lambda x: x[1], reverse=True))

# Set Seaborn style
sns.set_style("whitegrid")
plt.figure(figsize=(10, 6))

# Define color palette
colors = sns.color_palette("coolwarm", len(sorted_accuracy))

# Create bar plot
bars = plt.bar(sorted_accuracy.keys(), sorted_accuracy.values(), color=colors, edgecolor='black', linewidth=1.5)

# Add labels above bars
for bar in bars:
    plt.text(bar.get_x() + bar.get_width() / 2, 
             bar.get_height(), 
             f'{bar.get_height():.2f}', 
             ha='center', 
             va='bottom', 
             fontsize=12, 
             fontweight='bold')

# Customize labels and title
plt.xlabel('Models', fontsize=14, fontweight='bold', labelpad=10)
plt.ylabel('Accuracy', fontsize=14, fontweight='bold', labelpad=10)
plt.title('Model Accuracy Comparison (Sorted)', fontsize=16, fontweight='bold', pad=15)

# Improve readability with gridlines
plt.yticks(fontsize=12)
plt.xticks(fontsize=12, rotation=15)  # Rotate model names for better visibility
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show plot
plt.show()
```

```python
from sklearn.linear_model import SGDClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Initialize SGDClassifier with hinge loss (SVM-like)
sgd_model = SGDClassifier(loss='hinge', max_iter=1000, random_state=42)

# Train the SGDClassifier model
sgd_model.fit(X_train, y_train)

# Predict on test set
y_pred_sgd = sgd_model.predict(X_test)

# Calculate Accuracy
acc_sgd = accuracy_score(y_test, y_pred_sgd)
print(f'\nSGD Classifier Accuracy: {acc_sgd:.4f}')

# Classification Report
print(classification_report(y_test, y_pred_sgd))

# Confusion Matrix
plt.figure(figsize=(6, 4))
sns.heatmap(confusion_matrix(y_test, y_pred_sgd), annot=True, fmt='d', cmap='Blues')
plt.title('SGD Classifier Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```
